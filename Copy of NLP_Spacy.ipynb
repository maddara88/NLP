```python
text = "While a number of definitions of artificial intelligence (AI) have surfaced over the last few decades, John McCarthy offers the following definition in this 2004 paper (PDF, 106 KB) (link resides outside IBM),  It is the science and engineering of making intelligent machines, especially intelligent computer programs. It is related to the similar task of using computers to understand human intelligence, but AI does not have to confine itself to methods that are biologically observable."
```

```python
print(text)
```

```python
import spacy
```

```python
#download the english model
spacy.cli.download("en_core_web_sm")
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
nlp
```

```python
#Tokem
```

```python
doc = nlp(text)
```

```python
doc
```

```python
type(doc)
```

```python
for word in doc:
  print(word)
```

```python
sentences =[]
for sent in doc.sents:
  sentences.append(sent)
  print(sent)
```

```python
sentences
```

```python
sentences[0]
```

```python
sentences1 = sentences[1]
```

```python
for p,q in enumerate(sentences1):
  print(f" {p} \t {q}")
```

```python
for word in sentences1:
  print(f"{word}\t\t {word.is_punct}\t\t{word.is_currency}\t\t {word.is_stop}")
```

```python
for word in sentences[0].ents:
  print(f"{word.text}\t\t {word.label_}")
```

```python
spacy.displacy.render(sentences[0],style = "ent",jupyter = True)
```

```python
stop_words = []
non_stop_words = []
punch_words = []

for word in sentences[0]:
  if word.is_stop:
    stop_words.append(word)
  elif word.is_punct:
    punch_words.append(word)
  else:
    non_stop_words.append(word.lower_)
```
